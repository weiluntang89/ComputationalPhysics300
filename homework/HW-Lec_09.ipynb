This program is used for image processing. The original data file is a txt file, and it is plotted in the first figure. The second and third figures are the images after using Fourier transform. The second image is the amplitude of the Fourier transform, and the third image is the phase of the Fourier transform. 

# Imports

```python
import numpy as np
import matplotlib.pyplot as plt 
import os
```

# Main Loop

```python
arr = np.loadtxt("imagedata.txt")   #Read txt file and store in an numpy array

plt.figure(figsize = (8,8))
plt.imshow(arr, cmap = 'magma')     #original image 
plt.axis('off')
plt.colorbar()

#Save the first image 
plt.savefig("image1.png")
plt.show()

#Discrete Fourier's transform 
dft = np.fft.fft(arr)

#Plot the amplitude of the discrete Fourier's transform 
plt.figure(figsize = (8,8))
plt.imshow(abs(dft), cmap = 'magma')
plt.axis('off')
plt.colorbar()

#Save the second image 
plt.savefig("image2.png")
plt.show()

#Plot the phase of the discrete Fourier's transform 
plt.figure(figsize = (8,8))
plt.imshow(np.angle(dft), cmap = 'magma')
plt.axis('off')
plt.colorbar()

#Save the third image 
plt.savefig("image3.png")
plt.show()

#Print out images sizes 
print("The size of the original text file: ", 
      os.path.getsize("/Users/weiluntang12/imagedata.txt")* (1e-6), "MB")
print("The size of the first image: ", 
      os.path.getsize("/Users/weiluntang12/image1.png") * (1e-3), "kB")
print("The size of the second image: ", 
      os.path.getsize("/Users/weiluntang12/image2.png") * (1e-3), "kB")
print("The size of the third image: ", 
      os.path.getsize("/Users/weiluntang12/image3.png") * (1e-3), "kB")
```
