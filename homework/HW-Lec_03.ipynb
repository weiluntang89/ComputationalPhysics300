This program is to create an image of Mandelbrot set. 
The idea is to input complex numbers whose magnitudes are less than 2 into 
Mandelbrot set equation and calculate the numbers of iteration. 
Store all the numbers of iteration into a n x n array, and create a density plot.

# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
import time
```

# Function

```python
#Generate complex numbers and store them in a n x n array.
def complexArr(n):
    x = np.linspace(-2,2,n)
    y = np.linspace(-2,2,n)
    return [[complex(x[i],y[k]) for k in range(n)] for i in range(n)]

#This function is the formula of Mandelbrot set z_n+1 = (z_n)^2 + c.
#It takes a complex number and return the numbers of iteration.
#The maximum iteration in this function is 80.
#low iteration allows the program execute faster, and the image has more details 
#on the boundary of the Mandelbrot set.
def f(c):
    N = 0
    z = complex(0,0)
    while abs(z) <= 2 and N < 80:
        z = z**2 + c
        N += 1
    return N

#Store all numbers of iteration for corresponding complex in a n x n array. 
#The first parameter c is a complex array in this example. 
def mandelbrotSet(c, n):
    return np.transpose([[f(c[i][k]) for k in range(n)] for i in range(n)])
```

# Main Loop

```python
t0 = time.time()  #initial time 

n = 1000    #Image resolution. User can choose appropriate resolution by adjusting the 
            #value of n.
            #n = 2000 takes avg.15 sec, n = 1500 takes avg. 9 sec, and n = 1000 takes 
            #avg. 3 sec. 
            #Choose n = 1000 because there is no significant difference in resolution 
            #compared to n = 1500 and n = 2000.
            #The time to execute the program might be differet on a different device.
            
plt.figure(figsize = (35,35))  
plt.imshow(mandelbrotSet(complexArr(n),n), cmap = 'binary')
plt.colorbar()
plt.axis('off')
t1 = time.time()                #Final time 
print(f"Took {t1-t0} seconds")  #Show how many seconds it takes to run this program 

```
