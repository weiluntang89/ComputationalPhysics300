## Main Loop

```python
fib = {}                #Create a dictionary 
for n in range(1,301):  #User can change the range of the loop to any values
    if n == 1:
        num = 1
    elif n == 2:
        num = 1
    elif n > 2:
        num = fib[n-1] + fib[n-2]     #The nth value of Fibonacci is the sum of the previous two values
    fib[n]= num                       #Store nth value into the dictionary 
    print("The", n, "value of Fibonacci sequence:",num)
```
